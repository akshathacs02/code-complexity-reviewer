```python
from openai import OpenAI

# Add your API key
client = OpenAI(api_key="your_api")
```

```python
# Ask user for code
print("Paste your code below:")
print("Type END to finish\n")

lines = []

while True:
    line = input()

    if line == "END":
        break

    lines.append(line)

code = "\n".join(lines)

# Prompt
prompt = f"""
Analyze this code and give:

1. Time Complexity
2. Space Complexity
3. Improvements

Code:
{code}
"""

# Send to model
response = client.responses.create(
    model="gpt-5-mini",
    input=prompt
)

# Print result
print("\n===== RESULT =====\n")
print(response.output_text)
```

With UI integration

```python
import gradio as gr
from openai import OpenAI

# OpenAI client
client = OpenAI(api_key="YOUR_API_KEY")

# Function to analyze code
def analyze_code(user_code):

    prompt = f"""
    Analyze this code and give:

    1. Time Complexity (Big-O)
    2. Space Complexity
    3. Improvements

    Code:
    {user_code}
    """

    response = client.responses.create(
        model="gpt-5-mini",
        input=prompt
    )

    return response.output_text

# Gradio UI
with gr.Blocks() as app:

    gr.Markdown("# 📊 Code Complexity Reviewer")

    code_input = gr.Textbox(
        label="Paste your Python code here",
        lines=10,
        placeholder="Write or paste code..."
    )

    btn = gr.Button("Analyze Code")

    output = gr.Textbox(
        label="Analysis Result",
        lines=15
    )

    btn.click(
        fn=analyze_code,
        inputs=code_input,
        outputs=output
    )

# Run app
app.launch()
```
